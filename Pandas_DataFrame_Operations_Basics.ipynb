```python
import pandas as pd
import numpy as np

np.random.seed(0)

data = {"ID" : range(1,26),
        "Age": np.random.randint(18,60,size = 25),
        "Salary":np.random.randint(15000,50000,size = 25),
        "Department": np.random.choice(["HR","Finance","Marketing",'Developer'],size = 25),
        "Experience": np.random.randint(0,11,size = 25)}

Employees = pd.DataFrame(data)
```

```python
print(Employees)
```

```python
#Find Employees from Developer  Department

Employees_Dev = Employees[Employees["Department"] == 'Developer']

Employees_Dev
```

```python
#Arrange data in Descending order of age

sorted_age = Employees.sort_values(by = 'Experience', ascending = False)
```
