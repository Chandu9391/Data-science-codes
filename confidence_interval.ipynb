```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
glaxo_df = pd.read_csv("/content/glaxo_df.csv")
```

```python
glaxo_df
```

```python
beml_df = pd.read_csv("/content/beml_df.csv")
```

```python
beml_df
```

```python
beml_df_ci = stats.norm.interval(0.95,beml_df.gain.mean(),beml_df.gain.std())
```

```python
beml_df_ci
```

```python
np.round(beml_df_ci,5)
```
