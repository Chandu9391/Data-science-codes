```python
import re

# Function to convert duration to minutes
def convert_duration_to_minutes(duration):
    # Initialize hours and minutes
    hours = 0
    minutes = 0

    # Use regular expressions to find hours and minutes
    if 'h' in duration:
        hours = int(re.search(r'(\d+)h', duration).group(1))
    if 'm' in duration:
        minutes = int(re.search(r'(\d+)m', duration).group(1))

    # Convert total time to minutes
    return hours * 60 + minutes

# Apply the function to the Duration column
df_train['Duration_Minutes'] = df_train['Duration'].apply(convert_duration_to_minutes)

# Display the updated DataFrame
print(df_train[['Duration', 'Duration_Minutes']].head())
```

```python
#q7
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder

class PreprocessingPipeline(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.label_encoders = {}

    def fit(self, X, y=None):
        for col in X.select_dtypes(include=['object']).columns:
            self.label_encoders[col] = LabelEncoder().fit(X[col])
        return self

    def transform(self, X):
        X = X.drop_duplicates()

        X = X.dropna()

        X['Date'] = pd.to_datetime(X['Arrival_Time'], errors='coerce').dt.day
        X['Month'] = pd.to_datetime(X['Arrival_Time'], errors='coerce').dt.month
        X['Year'] = pd.to_datetime(X['Arrival_Time'], errors='coerce').dt.year

        for col, le in self.label_encoders.items():
            X[col] = le.transform(X[col])

        X = X.drop(columns=['Route', 'Arrival_Time', 'Dep_Time', 'Date_of_Journey', 'Year'], errors='ignore')

        return X

pipeline = Pipeline([
    ('preprocessing', PreprocessingPipeline())
])


test_data = pd.read_excel("Flight_Price_Test.xlsx")

test_data_transformed = pipeline.fit_transform(test_data)

test_data_transformed.head()
```
